```python
#Create a crime list of only violent crimes discard misdemeanors or nonviolent crimes
crime_list = [  'Kidnapping, abduction',
                'Statutory rape',
                'Drug, narcotic violations',
                'Theft from motor vehicle',
                'Simple assault',
                'Forcible sodomy',
                'Motor vehicle theft',
                'Murder, non-negligent',
                'Arson',
                'Forcible fondling',
                'Theft from building',
                'Human Trafficking/Involuntary Servitude',
                'Theft of motor vehicle parts or accessory',
                'Human Trafficking/Commercial Sex Act',
                'Aggravated Assault',
                'All other larceny',
                'Burglary, Breaking and Entering',
                'Destruction, damage, vandalism',
                'Negligent manslaughter',
                'Pocket-picking',
                'Purse-snatching',
                'Robbery',
                'Forcible rape',
            ]
```

```python
#Only pull crimes from the crime list into the df 
violent_crime_df = crime_2020_df[crime_2020_df['Offense_Type'].isin(crime_list)]
violent_crime_df.head()
```

```python
#Splits block numbers that are given in a range
violent_crime_df['Block_Range'] = violent_crime_df['Block_Range'].apply(lambda x: x.split('-')[0])
violent_crime_df
```

```python
#Create a new column called Street Address and use other columns to create a searchable street address
violent_crime_df['Street_Address'] = violent_crime_df['Block_Range'] + ' ' + violent_crime_df['Street_Name'] + ' ' + violent_crime_df['Street_Type']

violent_crime_df
```

```python
#Add a column called state and add TX to it for geolocation
violent_crime_df['State'] = 'TX'
violent_crime_df.head()
```

```python
#Create an output directory and save the file as an CSV to check 
dir = r"Output"
if not os.path.exists(dir):
    os.mkdir(dir)

violent_crime_df.to_csv('Output/violent_crime_cleaned.csv', index=False,)
```

```python
violent_crime_df['Address'] = violent_crime_df['Block_Range'] + ' ' + violent_crime_df['Street_Name'] + ' ' + violent_crime_df['Street_Type'] + ', ' + violent_crime_df['City'] + ', TX'

violent_crime_df.head()
```

```python
address_list = violent_crime_df['Address'].to_list()
print(len(address_list))
```

```python
violent_crime_df.to_csv('violent_crime_cleaned.csv', index=False)
```

```python
from geopy.geocoders import Nominatim

crime_lats = []
crime_lngs = []
counter = 0

for address in address_list[0:100]:
    try:
        geolocator = Nominatim(user_agent="Team1")
        location = geolocator.geocode(address)
# print(location.raw)
        print(f'{counter} {address} LatLng:{location.latitude, location.longitude}')
        print('----------------')
        crime_lats.append(location.latitude)
        crime_lngs.append(location.longitude)
    except:
        print(f'{counter} Not Found!')
        print('----------------')
        crime_lats.append('NaN')
        crime_lngs.append('NaN')
    counter+=1


```

```python
violent_crime_df['Latitude'] = crime_lats
violent_crime_df['Longitude'] = crime_lngs

violent_crime_df.head()
```

```python
violent_crime_df.to_csv('Lat_Lng.csv', index=False)
```
