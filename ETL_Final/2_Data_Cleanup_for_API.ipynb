```python
#Import dependancies
import pandas as pd
import os
import datetime as dt
import numpy as np
import requests
import csv
```

```python
#Read and view the data
crime_2020_cleaned = pd.read_csv("./Output/2020_crime_cleaned_lat_long1.csv")
crime_2020_cleaned
```

```python
crime_2020_cleaned['Offense_Type'].value_counts()
```

```python
#Group Crimes by type

#Theft
theft = ['Theft from motor vehicle', 'All other larceny', 'Motor vehicle theft', 'Theft of motor vehicle parts or accessory',
        'Theft from building', 'Purse-snatching', 'Pocket-picking', 'AutoTheft',]

crime_2020_cleaned['Offense_Type'][crime_2020_cleaned['Offense_Type'].isin(theft)] = 'Theft'

#Assault
assault = ['Aggravated Assault', 'Simple assault']

crime_2020_cleaned['Offense_Type'][crime_2020_cleaned['Offense_Type'].isin(assault)] = 'Assault'



#Sex Crime
sex_crime = ['Forcible rape','Statutory rape','Forcible fondling','Human Trafficking/Commercial Sex Act','Forcible fondling','Forcible sodomy',]

crime_2020_cleaned['Offense_Type'][crime_2020_cleaned['Offense_Type'].isin(sex_crime)] = 'Sex Crime'

#Burglary
burglary_robbery = ['Burglary, Breaking and Entering', 'Robbery',]

crime_2020_cleaned['Offense_Type'][crime_2020_cleaned['Offense_Type'].isin(burglary_robbery)] = 'Burglary/Robbery'

#Murder
murder = ['Murder, non-negligent','Negligent manslaughter',]

crime_2020_cleaned['Offense_Type'][crime_2020_cleaned['Offense_Type'].isin(murder)] = 'Murder/Manslaughter'

#Abduction
abduction = ['Kidnapping, abduction','Human Trafficking/Involuntary Servitude']

crime_2020_cleaned['Offense_Type'][crime_2020_cleaned['Offense_Type'].isin(abduction)] = 'Abduction'

#Narcotics
narcotics = ['Drug, narcotic violations']

crime_2020_cleaned['Offense_Type'][crime_2020_cleaned['Offense_Type'].isin(narcotics)] = 'Narcotics'


#Vandalism
vandalism = ['Destruction, damage, vandalism']

crime_2020_cleaned['Offense_Type'][crime_2020_cleaned['Offense_Type'].isin(vandalism)] = 'Vandalism'
```

```python
#Prep for api calls
crime_2020_cleaned['Street_Address'] = crime_2020_cleaned['Street_Address'].replace(' ', '+', regex=True)
crime_2020_cleaned['City'] = crime_2020_cleaned['City'].replace(' ', '+', regex=True)
crime_2020_cleaned['City'] = crime_2020_cleaned['City'].replace('HOUSTON', 'Houston', regex=True)

crime_2020_cleaned
```

```python
crime_2020_cleaned.to_csv('./Output/2020_crime_cleaned_lat_long_updated_prep.csv')
```
