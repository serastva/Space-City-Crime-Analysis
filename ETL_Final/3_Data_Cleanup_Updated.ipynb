```python
import pandas as pd
import csv
```

```python
crime_2020_cleaned = pd.read_csv("./Output/2020_crime_cleaned_lat_long_updated_prep_Tract.csv")
crime_2020_cleaned
```

```python
#Remove rows which contain a blank latitude
crime_2020_cleaned.dropna(subset = ['Latitude'], inplace = True) 
crime_2020_cleaned
```

```python
print(crime_2020_cleaned.shape)
```

```python
crime_2020_cleaned.Zip_Code.unique()
```

```python
#Remove rows which contain a blank zip_code
crime_2020_cleaned.dropna(subset = ['Zip_Code'], inplace = True) 
crime_2020_cleaned
```

```python
print(crime_2020_cleaned.shape)
```

```python
#Clean the Zip Codes
crime_2020_cleaned['Zip_Code'] = crime_2020_cleaned['Zip_Code'].replace('-0000', '', regex=True)
crime_2020_cleaned['Zip_Code'] = crime_2020_cleaned['Zip_Code'].replace('-3942', '', regex=True)
crime_2020_cleaned['Zip_Code'] = crime_2020_cleaned['Zip_Code'].replace('-4103', '', regex=True)
crime_2020_cleaned['Zip_Code'] = crime_2020_cleaned['Zip_Code'].replace('-4006', '', regex=True)
crime_2020_cleaned
```

```python
crime_2020_cleaned.Zip_Code.unique()
```

```python
print(crime_2020_cleaned.shape)
```

```python
#Remove zipcodes that were not in houston
zip_list = ['78582',
            '78501',
            '78557',
            '78596',
            '78418',
            '78414',
            '77979',
            '78941',
            '70601',
            '77701',
            '77566',
            '77486',
            '77448',
            '77356',
            '77378']

```

```python
crime_2020_cleaned = crime_2020_cleaned[~crime_2020_cleaned.Zip_Code.isin(zip_list)]
```

```python
crime_2020_cleaned.Zip_Code.unique()
```

```python
print(crime_2020_cleaned.shape)
```
